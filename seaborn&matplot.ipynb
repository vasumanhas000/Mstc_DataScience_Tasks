# Excersise on various Visualization Techniques

The iris dataset contains measurements for 150 iris flowers from three different species.


The three classes in the Iris dataset:
Iris-setosa (n=50)
Iris-versicolor (n=50)
Iris-virginica (n=50)
The four features of the Iris dataset:
sepal length in cm
sepal width in cm
petal length in cm
petal width in cm


Get the data
*Use seaborn to get the iris data by using: iris = sns.load_dataset('iris') *



```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

```

```python
from sklearn.datasets import load_iris
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
setosa = iris[iris['species']=='setosa']
sns.kdeplot( setosa['sepal_width'], setosa['sepal_length'],
                 cmap="plasma", shade=True, shade_lowest=False)
```

```python
tips=sns.load_dataset('tips')
tips['tip_pct']=tips['tip']*100/tips['total_bill']
g=sns.FacetGrid(data=tips,row='sex',col='time')
g.map(sns.distplot,'tip_pct',color='blue',bins=10,kde=False)
```

```python
tips=sns.load_dataset('tips')
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips);
```

```python
tips=sns.load_dataset('tips')
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind ="hex",
color="lightblue")
```
